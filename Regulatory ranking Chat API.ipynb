```python
import json
import os
from openai import OpenAI

# Set up your API key
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY")

# Function to call OpenAI API
def get_gpt_response(**kwargs):
    openai_client = OpenAI(api_key=OPENAI_API_KEY)
    return openai_client.chat.completions.create(**kwargs)

# Function to rank crops based on regulatory complexity
def rank_crops(crops):
    # Create the prompt to ask GPT to rank crops
    #prompt = f"Rank the following crops based on regulatory complexity, including factors such as compliance requirements, IP status, and sustainability concerns. Return your answer as JSON in the format {"answer": <list of ranked choices>}:\n\n"
    prompt = (
    f"Rank the following crops based on regulatory complexity, including factors such as compliance requirements, "
    f"IP status, and sustainability concerns. Return your answer as JSON in the format {{\"answer\": [<list of ranked choices>]}}:\n\n"
    )
    prompt += "\n".join(crops)

    # Call the GPT API using the get_gpt_response function
    response = get_gpt_response(
        model="gpt-4",  # Use the correct model name (gpt-4 or gpt-3.5)
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0,
        max_tokens=2000,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
    )
    
    # Extract the ranked crops from the response
    ranked_crops = json.loads(response.choices[0].message.content)
    #ranked_crops = response.choices[0].message.content  # Access the content as a string
    return ranked_crops

# List of crops to rank
crops = [
    "Cotton", "Cocoa", "Coffee (Arabica)", "Coffee (Robusta)", "Tea",
    "Coconut oil", "Groundnuts", "Palm oil", "Soybeans", "Rice", "Wheat",
    "Maize", "Bananas", "Barley", "Oats", "Sorghum", "Millet", "Potato",
    "Sweet Potato", "Cassava", "Sugarcane", "Tomato", "Cucumber", "Pepper",
    "Garlic", "Ginger", "Peanuts", "Almonds", "Walnuts", "Cashews", "Pistachios", "Vanilla"
]

# Get the ranked crops
ranked_crops = rank_crops(crops)

# Output the ranked crops
print("Ranked Crops Based on Regulatory Complexity:")
print(ranked_crops)
```

```python
import random
import os
import openai
from openai import OpenAI

# Creating randomized batches
def create_batches(crops, batch_size=5):
    random.shuffle(crops)  # Shuffle the crop list randomly
    return [crops[i:i + batch_size] for i in range(0, len(crops), batch_size)]

# Function to call OpenAI API
def get_gpt_response(**kwargs):
    openai_client = OpenAI(api_key=OPENAI_API_KEY)
    # Ensure the correct structure of messages
    for message in kwargs.get("messages", []):
        for k in ["function_call", "tool_calls"]:
            if k in message and message[k] is None:
                message.pop(k)
    # Call OpenAI's API
    response = openai_client.chat.completions.create(**kwargs)
    return response

# Function to rank crops based on regulatory complexity
def rank_crops(crops):
    # Create the prompt to ask GPT to rank crops
    #prompt = f"Rank the following crops based on regulatory complexity, including factors such as compliance requirements, IP status, and sustainability concerns. Return your answer as JSON in the format {"answer": <list of ranked choices>}:\n\n"
    prompt = (
    f"Rank the following crops based on regulatory complexity, including factors such as compliance requirements, "
    f"IP status, and sustainability concerns. Return your answer as JSON in the format {{\"answer\": [<list of ranked choices>]}}:\n\n"
    )
    prompt += "\n".join(crops)

    # Call the GPT API using the get_gpt_response function
    response = get_gpt_response(
        model="gpt-4",  # Use the correct model name (gpt-4 or gpt-3.5)
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0,
        max_tokens=2000,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
    )
    
    # Extract the ranked crops from the response
    ranked_crops = json.loads(response.choices[0].message.content)
    #ranked_crops = response.choices[0].message.content  # Access the content as a string
    return ranked_crops

# List of crops to rank
crops = [
    "Cotton", "Cocoa", "Coffee (Arabica)", "Coffee (Robusta)", "Tea",
    "Coconut oil", "Groundnuts", "Palm oil", "Soybeans", "Rice", "Wheat",
    "Maize", "Bananas", "Barley", "Oats", "Sorghum", "Millet", "Potato",
    "Sweet Potato", "Cassava", "Sugarcane", "Tomato", "Cucumber", "Pepper",
    "Garlic", "Ginger", "Peanuts", "Almonds", "Walnuts", "Cashews", "Pistachios", "Vanilla"
]

# Get the ranked crops
ranked_crops = rank_crops(crops)

# Output the ranked crops
print("Ranked Crops Based on Regulatory Complexity:")
print(ranked_crops)

# Create batches of 5 crops
batches = create_batches(crops, batch_size=5)

# Process each batch and rank them
for i, batch in enumerate(batches):
    print(f"\nRanking for Batch {i + 1}: {batch}")
    ranked_crops = rank_crops(batch)
    print(f"Ranked Batch {i + 1}: {ranked_crops}")
```
